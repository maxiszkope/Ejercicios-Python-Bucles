# Ejercicio 3

Dada una lista y un número, cree una función que devuelva dos parámetros:

1) Si está dicho número en esa lista. En dicho caso Devolverá `TRUE`. En caso contrario, `False`.

2) Si la longitud de la lista coincide con dicho número. En dicho caso devolverá `TRUE`. En caso contrario, `False`.

````python
lista = [2,5,6,7,8,9]
numero = 2

def comprobacion(l, n):
    """La funcion comprueba si el numero introducido se encuentra en la lista introducida y si dicho numero es igual a la longitud de la lista"""
    en_lista = n in l
    igualdad_longitud = len(l) == n
    return en_lista, igualdad_longitud 

comprobacion(lista,numero)
````

#Ejercicio 6

Dada una lista de palabras que defina y una palabra introducida por el usuario, analice si dicha palabra está en la lista.

En el caso de que la encuentre deberá avisar. 
Si no está, deberá añadirla a la lista. 

Use `lista_apalabras.append(palabra)` para añadir una palabra a la lista.

````python
palabras = ['perro','gato','mapache','raton','nutria']
termino = input('Ingrese palabra: ')

def analisis(pal,term):
    """Comprueba si la palabra esta en la lista y si no esta la añade al final"""
    if term in pal:
        print(f'La palabra {term} se encuentra en la lista')
    else:
        pal.append(term)
        print(f'La palabra {term} no estaba y se ha añadido al final de la lista')

analisis(palabras,termino)
print(f'Lista actualizada:{palabras}')


````

#Ejercicio 7
Cree una función para contar la cantidad de números negativos de una lista.


````python
lista = [-3,-5,7,9,2,45,-65,-4]

def contador_negativo(l):
    """Esta funcion indica la cantidad de numeros negativos en una lista"""
    return sum(1 for num in l if num < 0)

contador_negativo(lista)
````

# Ejercicio 8

Cree una función para contar la cantidad de números impares de una lista.


````python
lista = [1,2,3,4,5,6,7,8,9,12,13,14,16]

def contador_impar(l):
    """Esta funcion indica la cantidad de numeros impares en una lista"""
    return sum(1 for num in l if num % 2 != 0)

contador_impar(lista)
````

#Ejercicio 9

Dada una palabra, cuente el número de vocales que tiene.



````python
palabra = 'ornitorrinco'
def contador_vocales(p):
    """Esta funcion indica la cantidad de vocales en una palabra"""
    return sum(1 for voc in p.lower() if voc in 'aeiou')

contador_vocales(palabra)
````

````python
# Es muy común encontrar "While (True)".
# Aunque es muy peligroso que exista sin un buen control de lo que ocurre.

# Veamos un ejemplo para comprobar contraseñas

password = ""
while True: # <- Con True como condición, el bucle se ejecutaría permanentemente
    # "input" permite tomar valores por teclado. 
    password = input("Por favor ingrese la contraseña: ")
    if password == "secret":
        print("Gracias. Ha ingresado la contraseña correcta.")
        break # <- Pero con break podemos salir del bucle
    else:
        contador = 0
        if contador == 3:
            break
        contador += 1
        print("Lo sentimos, la contraseña es incorrecta - inténtelo de nuevo.")
        
        
````


Se pide (de forma independiente por cada apartado):

**1**) Pida un número por consola al usuario usando el siguiente comando:


```
numero = input("Por favor ingrese un número: ")
```
**2**) Controle que el número esté entre 1 y 9. Muestre un mensaje de error en caso de no complir dicha condición. El programa no debe continuar (no use `break` para realizar esta cuestión).

**3**) Encuentre aquellos factores superiores a dicho número. Para localizar un elemento de la matrix puede usar:

```
elemento = matrix[fila][columna]
```
*Pista*: Use bucles aninados, `range` y `len`.

**4**) Devuelva su matrix *traspuesta*.
[Matriz traspuesta](https://upload.wikimedia.org/wikipedia/commons/e/e4/Matrix_transpose.gif). Este punto puede realizarlo fuera de las condiciones realizadas anteriormente.

**Extra**) puede controlar si un usuario introduce otro valor (letra, símbolo u otro caracter) usando `isdigit`, `is_numeric` o `is_decimal`. ¿Cuál vendría mejor? 


**En todo momento trate de usar listas de comprensión, siempre que le sea posible.**

````python
matriz = [[1, 2],[3, 4],[5, 6],[7, 8]]

# 1

numero = int(input('Ingrese un número: '))

# 2

if numero >= 1 and numero <= 9:
    pass
else:
    print('El número debe estar comprendido entre 1 y 9')

# 3
    
for fila in range(len(matriz)):
    for columna in range(len(matriz[fila])):
        elemento = matriz[fila][columna]
        if elemento > int(numero):
            print(f"{elemento} es un factor mayor al número ingresado.")

# 4

# Dimensionar la nueva matriz (2*4) a partir de la otra (4*2)
matriz_trans = [[0] * len(matriz) for _ in range(len(matriz[0]))] 
# Traspaso de datos de filas a columnas
for fila in range(len(matriz)):
    for columna in range(len(matriz[fila])):
        matriz_trans[columna][fila] = matriz[fila][columna]

print(matriz_trans)
````
